```python
import keras
```

```python
from keras.models import Sequential
# Sequential is used to sequence the'input layer',
# 'hidden layer' and 'output lauer'in proper order.
```

# Overfitting

Overfitting is that we train our model too much on our training data that it learns everything and when it get new data it is not able to perform very good.



For example in the model 2 we can see that the model is learned with the training data in such a way that it completely separates red and yellow points. That is the model learned so much that it form a pattern shown in blue dotted line but when it get a new test data it just separate the red and yellow points incompletely with only the learned training pattern. That is model is performing bad. So, the model should have some error during training as shown in model 1.

```python
# Build Input Layer  & Hidden Layer One
model.add(Dense(units=30 , activation = 'relu' , kernel_initializer = 'uniform', input_dim=60))

# In our data there are 60 columns, so using 'input_dim=60' we will 
# create 60 nodes or neurons as inputs in the input layer.
# 'units=30' is used to create 30 nodes in the hidden layer.
# Hidden layer should not be more than the number of input layer. 
# 'Dense' is used because the connection between input and hidden 
# layer is very dense.
# ' kernel_initializer = 'uniform' is used to have uniform weight.
```

```python
model.add(Dropout(rate=0.1))
# 10% Dropout is used to prevent the overfitting
```
